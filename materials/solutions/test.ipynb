```python
import os

import ibis
from pins import board_connect
import pandas as pd
import pandera as pa

from rich import inspect
```

```python
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
board = board_connect(server_url='https://colorado.posit.co/rsc/')
```

```python
business_license_raw = board.pin_read("sam.edwardes/chicago-business-license-data-raw")
business_license_raw
```

```python
(business_license_raw["license_code"] + " - " + business_license_raw["license_description"]).value_counts().sort_values(ascending=False).head(50)
```

```python
food_inspection_raw = board.pin_read("sam.edwardes/chicago-food-inspection-data-raw")
food_inspection_raw
```

```python
df = pd.merge(
    food_inspection_raw,
    business_license_raw,
    how="inner",
    left_on="license_",
    right_on="license_id"
)

df
```

```python
(df["license_code"] + " - " + df["license_description"]).value_counts().sort_values(ascending=False).head(50)
```
